```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
npar = 100
pos = np.round(100 * np.random.rand(npar, 2), 0)
radius = np.round(10 * np.random.rand(npar, 1), 0)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
for x, y, r in zip(pos[:,0], pos[:,1], radius):
    circle = patches.Circle((x, y), r, edgecolor='w', facecolor='blue', alpha=0.5)
    ax.add_patch(circle)

circle = patches.Circle((110, 90), 10, edgecolor='w', facecolor='red', alpha=0.5)
ax.add_patch(circle)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Particle Position')
plt.grid(True)
plt.axis('equal')
plt.show()
```
