```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
delta = 0.01
x = np.arange(-5.0, 5.0, delta)
y = np.arange(-5.0, 5.0, delta)
X, Y = np.meshgrid(x, y)
```

```python
def function(X, Y):
    Z = (X - 0)**2 + (Y - 2)**2
    return Z

Z = function(X, Y)
fig, ax = plt.subplots(figsize=(4, 4))
cs = ax.contour(X, Y, Z)
ax.clabel(cs, cs.levels, inline=True, fontsize=10)
plt.axis('equal')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def function(X, Y):
    Z = (X - 2)
    return Z

Z = function(X, Y)
fig, ax = plt.subplots(figsize=(4, 4))
cs = ax.contour(X, Y, Z)
ax.clabel(cs, cs.levels, inline=True, fontsize=10)
plt.axis('equal')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
def function(X, Y):
    Z = X + Y
    return Z

Z = function(X, Y)
fig, ax = plt.subplots(figsize=(4, 4))
cs = ax.contour(X, Y, Z)
ax.clabel(cs, cs.levels, inline=True, fontsize=10)
plt.axis('equal')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
