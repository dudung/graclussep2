```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
import sys
import os
src_path = os.path.abspath(os.path.join('../..', 'src'))
if src_path not in sys.path:
    sys.path.insert(0, src_path)

from particles import generator as gen
xx, yy, rr = gen.create_cluster_238_circular(10.0)
```

```python
import random

def set_charge(xx, yy, rr, charge=1, num=92, mode='random'):
    n = min(len(xx), len(yy), len(rr))
    qq = [0] * n
    if mode == 'first':
        qq[0:num] = [charge] * num
    elif mode == 'last':
        qq[-num:] = [charge] * num
    elif mode == 'random':
        j = 0;
        for i in range(n):
            rnd = random.random()
            if j >= num:
                break
            if rnd > 0.5:
                qq[i] = charge;
                j += 1
    else:
        pass
    return qq
```

```python
qq = set_charge(xx, yy, rr, num=92, mode='first')
print(len(qq))
print(sum(qq))

fig, ax = plt.subplots(figsize=(8, 6))
for x, y, r, q in zip(xx, yy, rr, qq):
    if q == 0:
        color = '#8f8'
    else:
        color = '#f88'
    circle = patches.Circle((x, y), r, edgecolor='#888', facecolor=color, alpha=0.5, linewidth=2)
    ax.add_patch(circle)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Particle configuration (n = 146, p = 92, first), ')
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
qq = set_charge(xx, yy, rr, num=92, mode='last')
print(len(qq))
print(sum(qq))

fig, ax = plt.subplots(figsize=(8, 6))
for x, y, r, q in zip(xx, yy, rr, qq):
    if q == 0:
        color = '#8f8'
    else:
        color = '#f88'
    circle = patches.Circle((x, y), r, edgecolor='#888', facecolor=color, alpha=0.5, linewidth=2)
    ax.add_patch(circle)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Particle configuration (n = 146, p = 92, last), ')
plt.grid(True)
plt.axis('equal')
plt.show()
```
