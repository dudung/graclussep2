```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
import sys
import os
src_path = os.path.abspath(os.path.join('../..', 'src'))
if src_path not in sys.path:
    sys.path.insert(0, src_path)

from particles import generator as gen
from particles import modifier as mod
```

```python
xx, yy, rr = gen.create_cluster_238_circular(10.0)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10, 4))

qq = mod.set_charge(xx, yy, rr, num=92, mode='first')
for x, y, r, q in zip(xx, yy, rr, qq):
    if q == 0:
        color = '#8f8'
    else:
        color = '#f88'
    circle = patches.Circle((x, y), r, edgecolor='#888', facecolor=color, alpha=0.5, linewidth=2)
    ax1.add_patch(circle)

ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.grid(True)
ax1.axis('equal')
ax1.set_title('first')

qq = mod.set_charge(xx, yy, rr, num=92, mode='last')
for x, y, r, q in zip(xx, yy, rr, qq):
    if q == 0:
        color = '#8f8'
    else:
        color = '#f88'
    circle = patches.Circle((x, y), r, edgecolor='#888', facecolor=color, alpha=0.5, linewidth=2)
    ax2.add_patch(circle)

ax2.set_xlabel('X')
ax2.grid(True)
ax2.axis('equal')
ax2.set_title('last')

qq = mod.set_charge(xx, yy, rr, num=92, mode='random')
for x, y, r, q in zip(xx, yy, rr, qq):
    if q == 0:
        color = '#8f8'
    else:
        color = '#f88'
    circle = patches.Circle((x, y), r, edgecolor='#888', facecolor=color, alpha=0.5, linewidth=2)
    ax3.add_patch(circle)

ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.grid(True)
ax3.axis('equal')
ax3.set_title('random')

fig.suptitle('Particle positions (n = 146, p = 92)', fontsize=12)
plt.show()
```
