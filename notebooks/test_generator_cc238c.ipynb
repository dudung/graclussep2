```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
import sys
import os
src_path = os.path.abspath(os.path.join('../..', 'src'))
if src_path not in sys.path:
    sys.path.insert(0, src_path)

from particles import generator as gen
xx, yy, rr = gen.create_cluster_238_circular(10.0)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
for x, y, r in zip(xx, yy, rr):
    circle = patches.Circle((x, y), r, edgecolor='#484', facecolor='#afa', alpha=0.5, linewidth=2)
    ax.add_patch(circle)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Particle configuration (n = 238)')
plt.grid(True)
plt.axis('equal')
plt.show()
```
