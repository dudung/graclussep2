```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
npar = 20
pos = np.round(100 * np.random.rand(npar, 3), 0)
pos[:,2] = 0
vel = np.round(10 * np.random.rand(npar, 3), 0)
vel[:,2] = 0
mass = np.array([1] * npar)

Bfield = np.array([0, 0, 1])
charge = np.round(10 * np.random.rand(npar, 1) - 5, 0)

tsimul = np.linspace(0, 10, 1001)
dt = tsimul[1] - tsimul[0]
positions = [pos.copy()]
for t in tsimul:
    forces = np.cross(charge * vel, Bfield)
    acc = forces / mass[:,np.newaxis]
    vel = vel + acc * dt
    pos = pos + vel * dt

    positions.append(pos.copy())

positions = np.array(positions)
```

```python
plt.figure(figsize=(10, 6))
for i in range(npar):
    plt.plot(positions[:, i, 0], positions[:, i, 1], label=f'Particle {i+1}')

plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.title('Particle Trajectories')
plt.legend()
plt.xlim(0, 100)
plt.xticks(range(0, 110, 10))
plt.ylim(0, 60)
plt.grid(True)
plt.show()
```
