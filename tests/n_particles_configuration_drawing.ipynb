```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
npar = 238
diameter = 1
xx = []
yy = []
rr = []
for i in range(npar):
    if i == 0:
        x = 0
        y = 0
        xx.append(x)
        yy.append(y)
    elif 1 <= i and i <= 6:
        j = i - 1
        q = j * np.pi / 3.0
        x = diameter * np.cos(q)
        y = diameter * np.sin(q)
        xx.append(x)
        yy.append(y)
    elif 7 <= i and i <= 18:
        j = i - 7
        q = j * np.pi / 6.0
        x = 2 * diameter * np.cos(q)
        y = 2 * diameter * np.sin(q)
        xx.append(x)
        yy.append(y)
    elif 19 <= i and i <= 36:
        j = i - 19
        q = j * np.pi / 9.0
        x = 3 * diameter * np.cos(q)
        y = 3 * diameter * np.sin(q)
        xx.append(x)
        yy.append(y)
    elif 37 <= i and i <= 60:
        j = i - 37
        q = j * np.pi / 12.0
        x = 4 * diameter * np.cos(q)
        y = 4 * diameter * np.sin(q)
        xx.append(x)
        yy.append(y)
    elif 61 <= i and i <= 90:
        j = i - 61
        q = j * np.pi / 15.0
        x = 5 * diameter * np.cos(q)
        y = 5 * diameter * np.sin(q)
        xx.append(x)
        yy.append(y)
    else:
        pass
    rr.append(0.5 * diameter)
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
for x, y, r in zip(xx, yy, rr):
    circle = patches.Circle((x, y), r, edgecolor='#484', facecolor='#afa', alpha=0.5, linewidth=2)
    ax.add_patch(circle)

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Particle configuration')
plt.grid(True)
plt.axis('equal')
plt.show()
```
